```python
#import opencv
import cv2
```

```python
#load a pre-trained Haar Cascade classifier for detecting frontal faces
cascade_face = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
#capture video from the default camera (index 0)
cap = cv2.VideoCapture(0)
```

```python
#Create infinite loop for continuous capturing and processing video frames
while True:
    #Read a frame from the video capture object
    ret, img = cap.read()
    print(ret)
    
    #Convert the frame to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    #Detect faces in the grayscale frame using the pre-trained cascade classifier
    face = cascade_face.detectMultiScale(gray, 1.3, 5)
    
    #Draw rectangles around the detected faces
    for (x,y,w,h) in face:
        cv2.rectangle(img, (x,y),(x+w,y+h),(0,0,255),4)
        
        #Capture the region of interest (ROI) corresponding to the detected face
        face_roi = img[y:y+h, x:x+w]

        #Save the detected face to a file 
        cv2.imwrite('detected_face.jpg', face_roi)
    
    #Display the original color frame with face 
    cv2.imshow('img',img)
    
    #Wait for a key press, break the loop if the 'Esc' key (ASCII value 27) is pressed
    k = cv2.waitKey(30) & 0xff
    if k == 27:
        break

#Release the VideoCapture object
cap.release()

#Close all OpenCV windows
cv2.destroyAllWindows()
    
```
